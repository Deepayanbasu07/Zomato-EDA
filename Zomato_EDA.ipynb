## 1. City-Level Insights


### Number of Restaurants per City

```python
city_restaurant_count = final_df['City'].value_counts().head(10)  # Top 10 cities
city_restaurant_count.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.show()
```

### Average Rating per City

```python
city_avg_rating = final_df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
city_avg_rating.plot(kind='bar', figsize=(10, 6), color='coral')
plt.title("Top 10 Cities by Average Restaurant Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.show()
```

## 2. Cuisine Analysis


### Most Popular Cuisines


```python
# Count of cuisines
from collections import Counter

# Split and count cuisines as they are comma-separated
cuisines = final_df['Cuisines'].dropna().str.split(', ')
cuisine_count = Counter([cuisine for sublist in cuisines for cuisine in sublist])
cuisine_df = pd.DataFrame(cuisine_count.most_common(10), columns=['Cuisine', 'Count'])

# Plotting
cuisine_df.set_index('Cuisine').plot(kind='bar', figsize=(10, 6), color='lightgreen')
plt.title("Top 10 Most Popular Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.show()
```

### Average Rating by Cuisine


```python
# Average rating by cuisine
cuisine_ratings = final_df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
cuisine_ratings.plot(kind='bar', figsize=(10, 6), color='orange')
plt.title("Top 10 Cuisines by Average Rating")
plt.xlabel("Cuisine")
plt.ylabel("Average Rating")
plt.show()
```

## 3. Price Range Analysis


### Price Range Distribution


```python
price_range_dist = final_df['Price range'].value_counts().sort_index()
price_range_dist.plot(kind='bar', figsize=(8, 6), color='violet')
plt.title("Distribution of Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

### Average Cost for Two by City

```python
city_avg_cost = final_df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False).head(10)
city_avg_cost.plot(kind='bar', figsize=(10, 6), color='purple')
plt.title("Top 10 Cities by Average Cost for Two")
plt.xlabel("City")
plt.ylabel("Average Cost for Two")
plt.show()
```

## 4. Booking and Delivery Services


### Table Booking vs Rating


```python
table_booking_ratings = final_df.groupby('Has Table booking')['Aggregate rating'].mean()
table_booking_ratings.plot(kind='bar', figsize=(8, 6), color='skyblue')
plt.title("Average Rating for Restaurants with/without Table Booking")
plt.xlabel("Has Table Booking")
plt.ylabel("Average Rating")
plt.show()
```

### Online Delivery vs Rating


```python
online_delivery_ratings = final_df.groupby('Has Online delivery')['Aggregate rating'].mean()
online_delivery_ratings.plot(kind='bar', figsize=(8, 6), color='lightgreen')
plt.title("Average Rating for Restaurants with/without Online Delivery")
plt.xlabel("Has Online Delivery")
plt.ylabel("Average Rating")
plt.show()
```

## 5. Rating and Votes Analysis


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=final_df, hue='Price range', palette='coolwarm')
plt.title("Votes vs Ratings by Price Range")
plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
plt.show()
```
